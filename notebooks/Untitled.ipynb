```python
import pandas as pd
from collections import defaultdict
from itertools import chain
import pickle as pkl
```

```python
data=pd.read_csv("../data/Final_data.csv")
data
```

```python
file = open("../data/actors.pkl","wb")
actors = data["credits"].values
actors=list(actors)
actors = [str(i).split("-") for i in actors]
actors=list(chain.from_iterable(actors))
actors_set=set(actors)
actors_set = list(actors_set)
pkl.dump(actors_set,file)
file.close()
```

```python
file = open("../data/genres.pkl","wb")
genres = data["genres"].values
genres= list(genres)
genres = [str(i).split("-") for i in genres]
genres = list(chain.from_iterable(genres))
genres_set= set(genres)
genres_set=list(genres_set)
pkl.dump(genres_set,file)
file.close()
```
